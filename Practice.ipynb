```python
# CLASS ACTIVITY 2.2a - Add comments to code

# Original Price of Product been defined by a variable
Product_price = 250.00

# Discount Rate Provided for the Product
Discount_rate = 0.15

# Formular to Calculate Discount Amount
# where Product_price = 250.00 & Discount_rate = 0.15
Discount_amount = Product_price * Discount_rate

# Formular to Calculate Final Price of the Product
Final_price = Product_price - Discount_amount

print("Original price:", Product_price)
print("Discount:", Discount_amount)
print("Final price:", Final_price)
```
